```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# Question 1
transform1 = np.concatenate((np.linspace(0, 50, 50), np.linspace(100, 255, 100), np.linspace(150, 255, 106)), axis=0).astype(np.uint8)

emma = cv.imread(r"emma_gray.jpg", cv.IMREAD_GRAYSCALE)
assert emma is not None
emma_transformed = cv.LUT(emma, transform1)

fig, ax = plt.subplots(1, 2, figsize=(13, 6))
ax[0].plot(transform1)
ax[0].set_title("Intensity Transformation Function")
ax[1].imshow(emma_transformed, cmap = "gray", vmin=0, vmax=255)
ax[1].get_xaxis().set_visible(False)
ax[1].get_yaxis().set_visible(False)
ax[1].set_title("Transformed Image")

plt.show()
```

```python
# Question 2-1
transform2 = np.concatenate((np.linspace(255, 255, 150), np.linspace(255, 0, 50), np.linspace(0, 0, 56)), axis=0).astype(np.uint8)
brain = cv.imread("brain_proton_density_slice.png", cv.IMREAD_GRAYSCALE)

brain_transformed_1 = cv.LUT(brain, transform2)
fig, ax = plt.subplots(1, 2, figsize=(13, 6))
ax[0].plot(transform2)
ax[0].set_title("Intensity Transformation Function")
ax[1].imshow(brain_transformed_1, cmap = "gray", vmin=0, vmax=255)
ax[1].get_xaxis().set_visible(False)
ax[1].get_yaxis().set_visible(False)
ax[1].set_title("Transformed Image")

plt.show()
```
